#### citation matrix (directed)

```python
col_index2vid = {i: vid for i, vid in enumerate(set(pid_vid_dict.values()))}
vid2col_index = {vid: i for i, vid in col_index2vid.items()}
```

```python
out_deg = np.zeros(shape = (len(col_index2vid), len(col_index2vid)))
in_deg = np.zeros(shape = (len(col_index2vid), len(col_index2vid)))
```

```python
for line in mag.yield_one_line(ref_fname, delimiter=','):
    src, des = line
    s_vid, d_vid = pid_vid_dict[src], pid_vid_dict[des]
    if s_vid in vid2col_index:
        out_deg[vid2col_index[s_vid], vid2col_index[d_vid]] += 1
    if d_vid in vid2col_index:
        in_deg[vid2col_index[d_vid], vid2col_index[s_vid]] += 1
```

```python
out_deg_sum = np.sum(out_deg, axis=1)
vid_out_sum = {vid: out_deg_sum[i] for i, vid in col_index2vid.items()}
in_deg_sum = np.sum(in_deg, axis=1)
vid_in_sum = {vid: in_deg_sum[i] for i, vid in col_index2vid.items()}
```

```python
with open(mag.get_path('venue_citation_degrees.pickle'), 'wb') as file:
    pickle.dump((vid_out_sum, vid_in_sum), file)
```

```python
def norm_2d_array_in_place(a):
    for i in range(len(a)):
        total = np.sum(a[i])
        if total > 0:
            a[i] /= total
```

```python
norm_2d_array_in_place(in_deg)
norm_2d_array_in_place(out_deg)
```

```python
matrix = np.concatenate((out_deg, in_deg), axis = 1)
```

#### j2v model

```python
model = w2v.load_j2v('../../data/100feat_50minwords_10context_2016')
```

```python
vid2index = dict((vid, i) for i, vid in enumerate(model.index2word))
index2vid = {i: vid for vid, i in vid2index.items()}
```

```python
j2v_mat = model.wv.syn0 #model.syn0 #.astype('float64')
```

### Function

```python
def majority_vote(sample_pids):
    pred_venue = []
    true_venue = np.array([pid_vid_dict[pid] for pid in sample_pids])
    for i, pid in enumerate(sample_pids):
        c = Counter(paper_refp_dict[pid])
        pred_venue.append(c.most_common()[0][0])
    pred_venue = np.array(pred_venue)
    cn = np.sum(pred_venue == true_venue)
    return cn/len(sample_pids)
```

```python
def pred_pub_venue(sample_pids, matrix, vid2col_index, col_index2vid, power=1):
    sample_pids_avg_vec = np.zeros(shape=(len(sample_pids), matrix.shape[1]))
    
    for i, pid in enumerate(sample_pids):
        ix = []
        freq = []
        for vid, n in Counter(paper_refp_dict[pid]).items():
            if vid in vid2col_index:
                ix.append(vid2col_index[vid])
                freq.append(n)
        freq = np.power(freq, power)
        avg_vec = np.mean((matrix[ix, :].T * freq).T, axis=0)
        sample_pids_avg_vec[i] = avg_vec
        
    sims = cosine_similarity(sample_pids_avg_vec, matrix)
    pred_index = np.argsort(sims, axis=1)[:, -1]
    
    pred_venue = np.array([col_index2vid[pred_ix] for pred_ix in pred_index])
    true_venue = np.array([pid_vid_dict[pid] for pid in sample_pids])
    cn = np.sum(pred_venue == true_venue)
    return cn/len(sample_pids)
```

### Tuning the power

```python
n_sample = 10000
runs = 100
powers = list(range(1, 11))
pow_accuracy = {'citation': defaultdict(list), 'j2v': defaultdict(list)}
```

```python
sample_pids = random.sample(pids_2000, n_sample)
```

```python
majority_vote(sample_pids)
```

```python
for i in range(runs):
    sample_pids = random.sample(pids_2000, n_sample)
    for p in powers:
        j2v_acc = pred_pub_venue(sample_pids, j2v_mat, vid2index, index2vid, p)
        base_acc = pred_pub_venue(sample_pids, matrix, vid2col_index, col_index2vid, p)
        pow_accuracy['j2v'][p].append(j2v_acc)
        pow_accuracy['citation'][p].append(base_acc)
```

```python
for algo in pow_accuracy:
    print(algo)
    for power, li in pow_accuracy[algo].items():
        print(power, ':\t', np.mean(li))
```

### Comparison

```python
n_sample = 10000
runs = 100
power = 6
algo_accuracy = {'citation': [], 'j2v': [], 'vote': []}
```

```python
for i in range(runs):
    sample_pids = random.sample(pids_2000, n_sample)
    citation_acc = pred_pub_venue(sample_pids, matrix, vid2col_index, col_index2vid)
    j2v_acc = pred_pub_venue(sample_pids, j2v_mat, vid2index, index2vid)
    vote_acc = majority_vote(sample_pids)
    algo_accuracy['citation'].append(citation_acc)
    algo_accuracy['j2v'].append(j2v_acc)
    algo_accuracy['vote'].append(vote_acc)
```

```python
fig, ax = plt.subplots(figsize = (4, 3))
x_pos, labels, avgs, stds, counts = [], [], [], [], []

for i, (algo, corrs) in zip(range(len(algo_accuracy)), algo_accuracy.items()):
    x_pos.append(i)
    counts.append(len(corrs))
    labels.append(algo)
    avgs.append(np.mean(corrs))
    stds.append(np.std(corrs))

CIs = {'90': 1.645, '95': 1.96, '99': 2.576}
ax.bar(x_pos, avgs, yerr=CIs['95']*(stds/np.sqrt(counts)), align='center', alpha=0.5, ecolor='black', capsize=10)
ax.set_xticks(x_pos)
ax.set_xticklabels(labels)
ax.set_ylabel('Accuracy')
ax.yaxis.grid(True)
# ax.set_title("Combined")
plt.show()
fig.savefig('prediction_result.pdf', bbox_inches='tight', pad_inches=0)
```
